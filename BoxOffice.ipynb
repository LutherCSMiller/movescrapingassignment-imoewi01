```python
%matplotlib inline
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
r = requests.get("http://www.the-numbers.com/movie/records/All-Time-Domestic-Box-Office")
soup = BeautifulSoup(r.text, "lxml")
```

```python
trows = soup.find_all('tr')
```

# Looking at format of a single row

```python
one_row = trows[4]
bb = one_row.find_all('td')
bb
```

# Formatting the raw data as a dictionary {[Movies], [domestic gross]}
The first split call on the movie is to get rid of everything before summary">. 
The second call gets rid of everything after the last word in the title.


```python
rawdata = {'movie':[], 'gross':[], 'year':[]}
allrows = soup.find_all("tr")
for row in allrows[1:101]:
    el = row.find_all('td')
    movie = el[2].contents[0]
    movie = str(movie)
    movie = movie.split('summary">')[-1]
    movie = movie.split('</a')[0]
    gross = el[3].contents[0].replace('$','').replace(',', '')
    gross = float(gross)
    year = el[1].contents[0]
    rawdata['movie'].append(movie)
    rawdata['gross'].append(gross)
    rawdata['year'].append(year)
```

# Checking the output of the dictionary
Everything looks good here.

```python
for i in range(100):
    print(rawdata['movie'][i], rawdata['gross'][i], rawdata['year'][i])
```

```python
def get_adjusted_price(year, price):
    year = str(year)
    string = "https://www.statbureau.org/calculate-inflation-price-json?country=united-states&start={}/1/1&end=2016/08/01&amount={}".format(year, price)
    adjusted = requests.get(string)
    print(adjusted.json().replace('$','').replace(' ',''))
    if adjusted.status_code != 200:
        adjusted_price = math.nan
    adjusted = adjusted.json().replace('$','').replace(' ','')
    adjusted_price = float(adjusted)
    return adjusted_price
```
